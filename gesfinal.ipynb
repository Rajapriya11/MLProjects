Feature Extraction and Dimensionality Reduction

```python
import pandas as pd
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler

# Assuming you have a DataFrame named 'data' with all variables
# Assuming the target variable is named 'phase'

# Select the feature columns
feature_columns = df.columns[:-1]  # Exclude the last column (target variable)

# Select the target variable column
target_column = 'phase'

# Separate the features and target
X = df[feature_columns].values
y = df[target_column].values

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Calculate the correlation matrix
corr_matrix = np.corrcoef(X_scaled, rowvar=False)

# Define a correlation threshold
correlation_threshold = 0.7

# Find highly correlated variable pairs
highly_correlated_pairs = np.where(np.abs(corr_matrix) > correlation_threshold)
highly_correlated_variables = set()


# Extract the variable names of highly correlated pairs
for i, j in zip(highly_correlated_pairs[0], highly_correlated_pairs[1]):
    highly_correlated_variables.add(feature_columns[i])
    highly_correlated_variables.add(feature_columns[j])

# Select the highly correlated variables
X_highly_correlated = X_scaled[:, [df.columns.get_loc(var) for var in highly_correlated_variables]]

# Apply Linear Discriminant Analysis (LDA) for dimensionality reduction
n_components = min(X_highly_correlated.shape[1], len(np.unique(y)) - 1)
lda = LinearDiscriminantAnalysis(n_components=n_components)
X_reduced = lda.fit_transform(X_highly_correlated, y)

# Print the results before and after dimensionality reduction
print("Original shape:", X.shape)
print("Transformed shape:", X_reduced.shape)

```

# Machine Learning Classification Algorithms

Gradient Boost

```python
import optuna
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, roc_auc_score
from sklearn.preprocessing import label_binarize
from itertools import cycle

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.25, random_state=48)

# Define the objective function for Optuna
def objective(trial):
    params = {
        'loss': 'deviance',
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1),
        'n_estimators': trial.suggest_int('n_estimators', 100, 500),
        'max_depth': trial.suggest_int('max_depth', 3, 9),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10),
        'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2'])
    }

    # Train the model
    model = GradientBoostingClassifier(**params)
    model.fit(X_train, y_train)

    # Predict on the test set
    y_pred = model.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    return accuracy

# Evaluate the model before tuning
model_before = GradientBoostingClassifier(random_state=42)
model_before.fit(X_train, y_train)
y_pred_before = model_before.predict(X_test)
accuracy_before = accuracy_score(y_test, y_pred_before)
print("Accuracy before tuning:", accuracy_before)

# Perform Bayesian optimization
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=25)

# Extract the trial results
trials = study.trials
trial_numbers = [trial.number for trial in trials]
accuracy_scores = [trial.value for trial in trials]

# Plot the optimization history
plt.plot(trial_numbers, accuracy_scores, marker='o')
plt.xlabel('Trial Number')
plt.ylabel('Accuracy')
plt.title('Optimization History')
plt.grid(True)
plt.show()

# Get the best hyperparameters
best_params = study.best_params

# Create the best model with the tuned hyperparameters
best_model = GradientBoostingClassifier(random_state=42, **best_params)
best_model.fit(X_train, y_train)

# Evaluate the model on the test set
y_pred_after = best_model.predict(X_test)
accuracy_after = accuracy_score(y_test, y_pred_after)
print("Accuracy after tuning:", accuracy_after)

# Print confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_after))

# Print classification report
print("Classification Report:")
print(classification_report(y_test, y_pred_after))

# Compute ROC curve and ROC AUC for each class
y_scores = best_model.predict_proba(X_test)
y_test_bin = label_binarize(y_test, classes=best_model.classes_)
n_classes = len(best_model.classes_)
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_scores[:, i])
    roc_auc[i] = roc_auc_score(y_test_bin[:, i], y_scores[:, i])

# Plot ROC curve for each class
plt.figure()
colors = cycle(['blue', 'red', 'green'])  # Use different colors for each class
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2, label='Class %d (AUC = %0.2f)' % (i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - Gradient Boosting')
plt.legend(loc='lower right')
plt.show()
```

```python
import optuna
import lightgbm as lgb
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=48)

# Define the objective function for Optuna
def objective(trial):
    params = {
        'objective': 'multiclass',
        'num_class': 5,
        'boosting_type': 'gbdt',
        'metric': 'multi_logloss',
        'num_leaves': trial.suggest_int('num_leaves', 70, 100),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1),
        'max_depth': trial.suggest_int('max_depth', 7, 15),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'reg_alpha': trial.suggest_float('reg_alpha', 0.0, 1.5),
        'reg_lambda': trial.suggest_float('reg_lambda', 0.0, 1.5),
    }

    # Train the model
    model = lgb.LGBMClassifier(**params)
    model.fit(X_train, y_train)

    # Predict on the test set
    y_pred = model.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    return accuracy

# Evaluate the model before tuning
model_before = lgb.LGBMClassifier(random_state=42)
model_before.fit(X_train, y_train)
y_pred_before = model_before.predict(X_test)
accuracy_before = accuracy_score(y_test, y_pred_before)
print("Accuracy before tuning:", accuracy_before)

# Perform Bayesian optimization
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=20)

# Extract the trial results
trials = study.trials
trial_numbers = [trial.number for trial in trials]
accuracy_scores = [trial.value for trial in trials]

# Plot the optimization history
plt.plot(trial_numbers, accuracy_scores, marker='o')
plt.xlabel('Trial Number')
plt.ylabel('Accuracy')
plt.title('Optimization History')
plt.grid(True)
plt.show()
# Get the best hyperparameters
best_params = study.best_params
# Create the best model with the tuned hyperparameters
best_model = lgb.LGBMClassifier(random_state=50, **best_params)
best_model.fit(X_train, y_train)
# Evaluate the model on the test set
y_pred_after = best_model.predict(X_test)
accuracy_after = accuracy_score(y_test, y_pred_after)
print("Accuracy after tuning:", accuracy_after)
# Print confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_after))

# Print classification report
print("Classification Report:")
print(classification_report(y_test, y_pred_after))

# Compute ROC curve and ROC AUC for each class
y_scores = best_model.predict_proba(X_test)
y_test_bin = label_binarize(y_test, classes=best_model.classes_)
n_classes = len(best_model.classes_)
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_scores[:, i])
    roc_auc[i] = roc_auc_score(y_test_bin[:, i], y_scores[:, i])

# Plot ROC curve for each class
plt.figure()
colors = cycle(['blue', 'red', 'green'])  # Use different colors for each class
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2, label='Class %d (AUC = %0.2f)' % (i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)- Light GBM')
plt.legend(loc='lower right')
plt.show()

```
